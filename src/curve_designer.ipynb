# Curve designs

This file is meant as a playground to find interesting constraint functions.

```python
from sympy import *
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
import warnings
```

```python
var("x y", real=True)

sol = solve(y**4 + 3*x**2 - 5*y**3 + sin(5*x) - 1, y) # also interesting: x**4 + y**2 - x**2 - 1 = 0
```

```python
xspace = np.linspace(-10, 10, int(1e4), dtype=np.complex128)
sol_lambdas = [lambdify(x,s) for s in sol]

for sol_lambda in sol_lambdas:
    sol_vals = sol_lambda(xspace)
    mask = np.logical_not(np.isclose(sol_vals.imag, 0))
    plt.plot(np.ma.masked_array(xspace, mask), np.ma.masked_array(sol_vals, mask))
    
plt.show()
```
