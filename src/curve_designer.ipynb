# Curve designs

This file is meant as a playground to find interesting constraint functions.

```python
from sympy import *
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
import warnings

import geometry
```

```python
var("x y", real=True)

g = sinh(x)**2 + sinh(y)**2 - 1 # x**4 + y**2 - 2*x**2 +2*x**3 - 1 # x**4 + y**2 - x**2 + 0.1*sin(20*x) - 1 # also interesting: x**4 + y**2 - x**2 - 1 = 0

curve = geometry.CurveConstraint(g, x, y, inside="negative", xlim=(-5,5))
fig, ax = plt.subplots()
ax.set_aspect("equal")
curve.plot(ax)
plt.show()
```
