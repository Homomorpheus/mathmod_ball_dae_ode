# Donut shape

```python
import sympy
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output, Video, HTML, update_display
import time
from typing import Literal

import sys
sys.path += ["../src", "./src"]

import solvers
import geometry
import animation
```

```python
sympy.var("x y", real=True)

curve = geometry.CurveConstraint((x**2 + y**2)**2 - (x**2 + y**2) + 0.1, x, y, inside="negative", xlim=(-1.2,1.2))
```

```python
gravity = lambda q, v: np.array([0, -3])

ball = geometry.Ball(curve, mass=1, gravity=gravity, curve_friction=0.1, normal_cor=0.5, tangential_cor=0.9, bounce_threshold=0.05)
```

```python
t_end = 6
steps = int(80*t_end)

status = widgets.HTML("-")
display(status)

anim = animation.BallAnimation(curve, [ball], steps)

def callback(step, qnew, vnew, error, mode: Literal["dae", "ode"]):
    
    status.value = F'<b>step {step} of {steps}</b> <progress value="{step}" max="{steps}"> </progress>'
    anim.update(qnew, vnew, 0, step-1, mode)

ball.simulate(steps, t_end, q0=(0.1, 0.25), v_up=0, callback=callback) # also interesting: (-3, 5), 0; (4,5), -1; (2,2), -1; (0,5), 0
```

```python
# play
HTML(anim.render(1/60))
```
